```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


#  
data_path = '/content/week5_test.csv'
data = pd.read_csv(data_path)

#  
variables = ['GyroX', 'GyroY', 'GyroZ', 'Temperature', 'Heartbeat']

#   
correlation = data[variables].corr()

#   
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#   
plt.figure(figsize=(12, 10))
for i, var in enumerate(variables):
    plt.subplot(3, 2, i+1)
    sns.lineplot(data=data, x='time', y=var)
    plt.title(var)
    plt.xlabel('Time')
    plt.ylabel(var)
plt.tight_layout()
plt.show()
```

```python
def plot_activity(activity, data):
    fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(nrows=5, figsize=(15, 7), sharex=True)
    plot_axis(ax0, data['UserCode'], data['time'], data['GyroX'], 'X-Axis')
    plot_axis(ax1, data['UserCode'], data['time'], data['GyroY'], 'Y-Axis')
    plot_axis(ax2, data['UserCode'], data['time'], data['GyroZ'], 'Z-Axis')
    plot_axis(ax3, data['UserCode'], data['time'], data['Temperature'], 'Temperature')
    plot_axis(ax4, data['UserCode'], data['time'], data['Heartbeat'], 'Heartbeat')
    plt.subplots_adjust(hspace=0.2)
    fig.suptitle(activity)
    plt.subplots_adjust(top=0.90)
    plt.show()

def plot_axis(ax, user, x, y, title):
  for i in range(len(x)-1):
    ax.plot([x.iloc[i], x.iloc[i+1]],[y.iloc[i], y.iloc[i+1]], color=u2c[user.iloc[i]])
  ax.set_title(title)
  ax.xaxis.set_visible(False)
  ax.set_ylim([min(y) - np.std(y), max(y) + np.std(y)])
  ax.set_xlim([min(x), max(x)])
  ax.grid(True)
```

```python
from itertools import cycle

users = data['UserCode'].value_counts().index
colors = 'bckgmwry'
u2c = {k:v for (k, v) in zip(users, cycle(colors))}
u2c
```

```python
for activity in data['Label'].value_counts().index:
    mask = (data['Label'] == activity)
    data_for_plot = data[mask][:200]
    plot_activity(activity, data_for_plot)
```
